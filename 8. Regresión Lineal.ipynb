# Análisis predictivo

## Usando Algoritmo SVM  (support vector machine)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.svm import SVR
```

* Mientras mayor sea la dispersión de los datos, mayor será el % de error

```python
dispersion = 7
```

```python
X = np.sort(6*np.random.rand(200,1), axis=0)
Y = np.sin(X).ravel()
Y[::5] += dispersion *(0.5-np.random.rand(40))
```

```python
plt.scatter(X,Y, color="darkorange", label="data")
```

```python
#Se crea el modelo
C = 1e3
svr_lin = SVR(kernel="linear", C=C)    #Modelo lineal
svr_rbf = SVR(kernel="rbf", C=C, gamma=0.1)  #modelo radial
svr_poly = SVR(kernel="poly", C=C, degree=3) #modelo polinomial
```

```python
#Se ajusta el modelo
svr_lin.fit(X,Y)
svr_rbf.fit(X,Y)
svr_poly.fit(X,Y)
#Predicción
y_lin = svr_lin.predict(X)
y_rbf = svr_rbf.predict(X)
y_poly = svr_poly.predict(X)
```

```python
#Coeficiente de error
print( r2_score(Y, y_lin) )
print( r2_score(Y, y_rbf) )
print( r2_score(Y, y_poly) ) 
```

```python
#Gráfica
lw = 2
plt.figure(figsize=(8,6))
plt.scatter(X,Y, color="darkorange", label="data")
plt.plot(X,y_lin, color="navy", lw = lw, label="SVM Lineal")
plt.plot(X,y_rbf, color="c", lw = lw, label ="SVM Radial")
plt.plot(X,y_poly, color="cornflowerblue", lw=lw, label="SVM Polinomial")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Support Vector Regression")
plt.legend()
```

* Las SVM funcionan MUY bien con grandes datos
